Create 5 boxes of different sizes and colors in one image. it is recommended to use random numbers!

```python
import random
rand_range = np.random.randint(600, size=(5,4))
#rand_color = np.random.randint(255, size=(5,3))
img_9 = img2.copy()

i = 0
while i < 5:
  rand_clr1 = random.randint(0,255)
  rand_clr2 = random.randint(0,255)
  rand_clr3 = random.randint(0,255)
  cv.rectangle(img_9,(rand_range[i][0],rand_range[i][1]),(rand_range[i][2],rand_range[i][3]),color=(rand_clr1,rand_clr2,rand_clr3),thickness=10)
  i += 1
plt.imshow(img_9)
```
