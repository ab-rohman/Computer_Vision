```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Step 2 (face det)
face_cascade = cv.CascadeClassifier('/content/drive/MyDrive/Colab Notebooks/Images_PCVK/Images/haarcascades/haarcascade_frontalface_alt.xml')

# Presiden Face
presiden = cv.imread('/content/drive/MyDrive/Colab Notebooks/Images_PCVK/Images/facedet/jokowi.jpg')
presiden1 = cv.imread('/content/drive/MyDrive/Colab Notebooks/Images_PCVK/Images/facedet/jokowi.jpg')
roi_wajah1 = face_cascade.detectMultiScale(presiden1) 

roi_face = face_cascade.detectMultiScale(presiden) 
for(x,y,w,h) in roi_face: 
 cv.rectangle(presiden,(x,y),(x+w,y+h),(0,255,0),3) 
plt.figure(figsize = (10,10)) 

f, axarr = plt. subplots(1,2,figsize=(15,5))
plt.subplot(121),plt.imshow(cv.cvtColor(presiden1, cv.COLOR_BGR2RGB))
axarr[0].imshow(cv.cvtColor(presiden1, cv.COLOR_BGR2RGB))
plt.subplot(122),plt.imshow(cv.cvtColor(presiden, cv.COLOR_BGR2RGB))
axarr[1].imshow(cv.cvtColor(presiden, cv.COLOR_BGR2RGB))
plt.show()
```

```python
#Step 2 (face det)
# Ibu Kartini Face
ibu_kartini = cv.imread('/content/drive/MyDrive/Colab Notebooks/Images_PCVK/Images/facedet/kartini.jpg')
ibu_kartini1 = cv.imread('/content/drive/MyDrive/Colab Notebooks/Images_PCVK/Images/facedet/kartini.jpg')
roi_wajah1 = face_cascade.detectMultiScale(ibu_kartini1) 

roi_face = face_cascade.detectMultiScale(ibu_kartini) 
for(x,y,w,h) in roi_face: 
 cv.rectangle(ibu_kartini,(x,y),(x+w,y+h),(0,255,0),3) 
plt.figure(figsize = (10,10)) 

f, axarr = plt. subplots(1,2,figsize=(15,5))
plt.subplot(121),plt.imshow(cv.cvtColor(ibu_kartini1, cv.COLOR_BGR2RGB))
axarr[0].imshow(cv.cvtColor(ibu_kartini1, cv.COLOR_BGR2RGB))
plt.subplot(122),plt.imshow(cv.cvtColor(ibu_kartini, cv.COLOR_BGR2RGB))
axarr[1].imshow(cv.cvtColor(ibu_kartini, cv.COLOR_BGR2RGB))
plt.show()
```
