```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Step 3
img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Images_PCVK/Images/j.png', 1)
kernel = np.ones((5, 5), 'uint8')
dilate_img = cv2.dilate(img, kernel, iterations=1)

fig, frame = plt.subplots(1,2)
frame[0].imshow(img)
frame[1].imshow(dilate_img)
```

```python
#Step 4
erosion_img = cv2.erode(img, kernel, iterations = 1)
fig, frame = plt.subplots(1,2)
frame[0].imshow(img)
frame[1].imshow(erosion_img)
```

```python
#Step 5
morph_dilation = cv2.morphologyEx(img, cv2.MORPH_DILATE, kernel)
fig, frame = plt.subplots(1,2)
frame[0].imshow(img)
frame[1].imshow(morph_dilation)
```

```python
#Step 6
morph_erosion = cv2.morphologyEx(img, cv2.MORPH_ERODE, kernel)
fig, frame = plt.subplots(1,2)
frame[0].imshow(img)
frame[1].imshow(morph_erosion)
```

```python
#Step 7
kernel2 = np.ones((7, 7), 'uint8')
morph_opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel2)
fig, frame = plt.subplots(1,2)
frame[0].imshow(img)
frame[1].imshow(morph_opening)
```
